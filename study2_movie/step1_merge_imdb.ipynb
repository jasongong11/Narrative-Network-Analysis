```python
import pandas as pd
```

```python
imdb_all_data = pd.read_csv("title.basics.tsv", sep="\t")
imdb_all_data = imdb_all_data[imdb_all_data.titleType=="movie"]
imdb_rating_data = pd.read_csv("title.ratings.tsv", sep="\t")
imdb_all_data = imdb_all_data.merge(imdb_rating_data, how="left", on="tconst")
imdb_crew_data = pd.read_csv("title.crew.tsv", sep="\t")
imdb_all_data = imdb_all_data.merge(imdb_crew_data[["tconst", "writers"]], how="left", on="tconst")
```

```python
meta_df = pd.read_csv("meta_df_with_imdb_id.csv")
imdb_all_data = imdb_all_data.rename({"genres": "imdb_genres"}, axis=1)
meta_with_imdb_df = meta_df.merge(imdb_all_data, how="left", left_on="imdb_id_final", right_on="tconst")
```
